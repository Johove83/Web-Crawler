```python
import requests
from bs4 import BeautifulSoup
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="ResultsContainer")
#print(f"{results} + \n\n\n\n")

#results.prettify()

job_elements = results.find_all("div", class_="card-content")

for job_element in job_elements:
    print(job_element.find("h2", class_="title").text.strip())
    print(job_element.find("h3", class_="company").text.strip())
    print(job_element.find("p", class_="location").text.strip())
    print("\n\n")
    
"""could also have named variables as follows
    title_element - job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    etc 
    print
"""
```
